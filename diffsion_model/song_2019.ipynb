```python
## Configuration
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import plotly.express as px
import plotly.graph_objects as go 
```

```python
## Noisy data distribution figure 
# Creating dataset
N = 2000
x = np.random.uniform(low=-3.0, high=3.0, size=N)
y = np.random.uniform(low=-3.0, high=3.0, size=N)
z = np.cos(0.2 * x) + np.sin(0.1 * y) 
```

```python
# for creating a responsive plot
%matplotlib widget

fig = plt.figure(figsize = (14, 9))
ax = Axes3D(fig, auto_add_to_figure=False)
fig.add_axes(ax)

ax.scatter(x, y, z, color='green',  s = 0.7)
plt.show()
```

```python
## Interactive plot with plotly
data = go.Scatter3d(x = x, y = y, z = z, mode = 'markers', marker = dict(size = 0.7, color = z, colorscale='Viridis', opacity = 0.8))
fig = go.Figure(data = data)

fig.update_layout(margin = dict(l=0, r=0, b=0, t=0))
fig.write_html("real_data.html")
fig.show()
```

```python
## Noisy data distribution figure 2
# Creating dataset
scale = 0.2
x2 = x + np.random.normal(loc=0.0, scale = scale, size=N) 
y2 = y + np.random.normal(loc=0.0, scale = scale, size=N) 
z2 = z + np.random.normal(loc=0.0, scale = scale, size=N) 
```

```python
# for creating a responsive plot
%matplotlib widget

fig = plt.figure(figsize = (14, 9))
ax = Axes3D(fig, auto_add_to_figure=False)
fig.add_axes(ax)

ax.scatter(x2, y2, z2, color='green',  s = 0.7)
plt.show()
```

```python
## Interactive plot with plotly
data = go.Scatter3d(x = x2, y = y2, z = z2, mode = 'markers', marker = dict(size = 0.7, color = z, colorscale='Viridis', opacity = 0.8))
fig = go.Figure(data = data)

fig.update_layout(margin = dict(l=0, r=0, b=0, t=0))
fig.write_html("noisy_data.html")
fig.show()
```
